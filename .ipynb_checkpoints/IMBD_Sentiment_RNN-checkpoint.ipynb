```python
nlp = spacy.load('en_core_web_sm')
stopwords = nlp.Defaults.stop_words
```

```python
# extracting text from tags 
def strip_html(text):
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text()

# removing square brackets
def remove_between_square_brackets(text):
    return re.sub('\[[^]]*\]', '', text)

# lemmatization and removal of stop words
def lemma(text):
    doc = nlp(text.lower())
    lemmas = [token.lemma_ for token in doc 
          if (token.lemma_.isalnum() or token.lemma_ in ['.', '!', '?'] ) and token.lemma_ not in stopwords]
    text = ' '.join(lemmas)
    return text

# cleaning text
def clean(text):
    text = strip_html(text)
    text = remove_between_square_brackets(text)
    text = lemma(text)
    return text
```

```python
IMDB_Dataset_Clean = pd.read_csv('IMDB_Dataset_Clean.csv')
```

```python
sentences = []
_ = [sentences.append(x.split()) for x in IMDB_Dataset_Clean['review']]
```

```python
t = time.time()

model = gensim.models.Word2Vec(
    sentences=sentences,
    window=3,
    min_count=1,
    vector_size=100, 
    workers=12)

print('Time to build vocab: {} mins'.format(round((time.time() - t) / 60, 2)))
```

```python
word2vecDict = {}
for index, word in enumerate(model.wv.index_to_key):
#     print(f"word #{index}/{len(model.wv.index_to_key)} is {word}")
#     print(model.wv.vectors[index])
    word2vecDict[word] = model.wv.vectors[index]
```

```python
list(word2vecDict.keys())[:10]
```

```python
movie_reviews_vectorized = []
for review in IMDB_Dataset_Clean['review']:
    temp = []
    for word in review.split():
        temp.append(word2vecDict[word])
    movie_reviews_vectorized.append(np.array(temp))
```

```python
# encoding sentiment
le = preprocessing.LabelEncoder()
movie_reviews_sentiment = le.fit_transform(IMDB_Dataset_Clean['sentiment'])
```

```python
# splitting data
X_train, X_test, y_train, y_test = train_test_split(movie_reviews_vectorized, movie_reviews_sentiment, test_size=0.33, random_state=42)
```

```python
import torch.nn as nn
```

```python
class RNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers):
        super(RNN, self).__init__()
        
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        
        self.rnn = nn.RNN(input_size, hidden_size, num_layers=num_layers)
        self.dropout = nn.Dropout(0.1)
        self.fc = nn.Linear(hidden_size, output_size)
        self.softmax = nn.LogSoftmax(dim=1)
    
    def forward(self, input, hidden):
        output, hidden = self.rnn(input, hidden)
        output = self.dropout(output)
        output = output.squeeze()[-1].unsqueeze(0)
        output = self.fc(output)
        output = self.softmax(output)
        return output, hidden

    def initHidden(self):
        return torch.zeros(self.num_layers*1, 1, self.hidden_size)
```

```python
input_size = len(X_train[0][0])
output_size = 2

hidden_size = 128
num_layers = 2

rnn = RNN(input_size, hidden_size, output_size, num_layers)

# criterion = nn.CrossEntropyLoss()
criterion = nn.NLLLoss()

lr = 0.001
# optimizer = torch.optim.Adam(rnn.parameters(), lr=lr)
optimizer = torch.optim.SGD(rnn.parameters(), lr=lr)
```

```python
def train(X_train, y_train):
    hidden = rnn.initHidden()
    rnn.zero_grad()
    output, hidden = rnn(X_train, hidden)
    loss = criterion(output, y_train)
    loss.backward()
#     nn.utils.clip_grad_norm_(rnn.parameters(), max_norm=2.0)
    optimizer.step()

    return output, loss.item()
```

```python
def getLabel(output):
    value_tensor, index_tensor = output.topk(1)
    return index_tensor[0].item()
```

```python
def getElapsedTime(start_time):
    current_time = time.time()
    s = current_time - start_time
    m = math.floor(s / 60)
    s -= m * 60
    return '{}m {:.0f}s'.format(m, s)
```

```python
n_iters = len(X_train)
print_every = 5000
plot_every = 1000

start_time = time.time()

all_losses = []

for epoch in range(3):
    epoch_loss = []
    current_loss = 0
        
    for i in range(n_iters):
        
        output, loss = train(torch.tensor(X_train[i]).unsqueeze(1), torch.tensor([y_train[i]], dtype=torch.long))
        current_loss += loss
    
        if i % print_every == 0:
            guess = getLabel(output)
            if guess == y_train[i]:
                guess = "Correct! It's {}".format(guess)  
            else:
                guess = 'Wrong! Guessed {} but should be: {}'.format(guess, y_train[i])
        
            print('Epoch %d Sample %d | %d%% | (%s) | Loss: %.4f | %s' % (epoch, i, i / n_iters * 100, getElapsedTime(start_time), loss, guess))

        # Add current loss avg to list of losses
        if i % plot_every == 0:
            epoch_loss.append(current_loss / plot_every)
            all_losses.append(current_loss / plot_every)
        
            current_loss = 0
    
    
    plt.title('Epoch {} Loss'.format(epoch))
    plt.plot(epoch_loss)
    plt.show()
    plt.title('All Losses')
    plt.plot(all_losses)
    plt.show()
```

```python
plt.plot(all_losses)
plt.show()
```

```python
def evaluate(X_test):
    hidden = rnn.initHidden()
    for i in range(len(X_test)):
        output, hidden = rnn(X_test[i].unsqueeze(0), hidden)
    return output
```

```python
# testing
n_iters = len(X_test)

correct = 0
for i in range(n_iters):
    with torch.no_grad():
        output = evaluate(X_test[i])
        
        output = getLabel(output)

        if output == y_test[i]:
            print('Correct!')
            correct+=1
        else:
            print('Incorrect!')
            
print('Accuracy of Model: {:.2f}'.format(correct / n_iters))
```
